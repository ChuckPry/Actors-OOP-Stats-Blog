3. This is the class creation cell.  I defne the attributes of the class.  If you notice I instanitated a local variable list so I can just manipulate the actual self variable.  I created a function inside the class for raw score to tally all scores from the four categories.  I then created the callbackchance function so I can calculate everyone's score with the weights.

```python
actlist = []
class Actors:
    def __init__(self, actor_id, abdominals, clark_gable_jaw, audition_params, perceived_age):
        self.actor_id = actor_id
        self.abdominals = abdominals
        self.clark_gable_jaw = clark_gable_jaw
        self.audition_params = audition_params
        self.perceived_age = perceived_age
        actlist.append(self)
    
    
    def callbackchance(self):
        eight_pack = (self.abdominals * 33) * .4
        params = (self.audition_params * 25) * .25
        jaw = (self.clark_gable_jaw * 50) * .15
        age = (self.perceived_age * 50) * .20
        new_score = eight_pack + params + jaw + age
        return new_score 
    
    



        
```
